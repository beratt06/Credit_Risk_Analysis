```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from pathlib import Path
```

```python
base_path = Path.cwd()

csv_path = base_path / "credit_risk_dataset.csv"

df = pd.read_csv(csv_path)
df = df.dropna()
```

```python
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")
sns.set_palette("Set2")

# Sıralı çubuk grafik çizimi
ax = sns.countplot(data=df, x='loan_intent', order=df['loan_intent'].value_counts().index)

# Başlık ve etiketler
plt.title("Kredi Türüne Göre Başvuru Dağılımı", fontsize=14)
plt.xlabel("Kredi Türü", fontsize=12)
plt.ylabel("Başvuru Sayısı", fontsize=12)
plt.xticks(rotation=45)

# Çubuk üstüne değer etiketleme
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}',
                (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")
sns.set_palette("pastel")

# Ortalama faiz oranlarını gösteren barplot
ax = sns.barplot(data=df, x='loan_intent', y='loan_int_rate', estimator='mean')

# Başlık ve etiketler
plt.title("Kredi Türüne Göre Ortalama Faiz Oranı (%)", fontsize=14)
plt.xlabel("Kredi Türü", fontsize=12)
plt.ylabel("Ortalama Faiz (%)", fontsize=12)
plt.xticks(rotation=45)

# Her çubuğun üstüne ortalama değer yazdırma
for p in ax.patches:
    height = p.get_height()
    if not np.isnan(height):
        ax.annotate(f'{height:.2f}%',
                    (p.get_x() + p.get_width() / 2., height),
                    ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")

# Kredi notuna göre onay durumu dağılımı
ax = sns.countplot(
    data=df,
    x='loan_grade',
    hue='loan_status',
    order=sorted(df['loan_grade'].unique()),
    palette="Set1"
)

# Başlık ve eksen etiketleri
plt.title("Kredi Notuna Göre Onay Durumu", fontsize=14)
plt.xlabel("Kredi Notu (Grade)", fontsize=12)
plt.ylabel("Başvuru Sayısı", fontsize=12)
plt.legend(title="Durum", title_fontsize=11, fontsize=10)

# Sayı etiketleri ekleme
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{height}',
                    (p.get_x() + p.get_width() / 2., height),
                    ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")

# Histogram + KDE
ax = sns.histplot(df['cb_person_cred_hist_length'], bins=15, kde=True, color="skyblue")

# Başlık ve eksenler
plt.title("Kredi Geçmişi Süresi Dağılımı", fontsize=14)
plt.xlabel("Yıl", fontsize=12)
plt.ylabel("Kişi Sayısı", fontsize=12)

# Sayı etiketleri yalnızca KDE kapalı olduğunda güvenlidir,
# fakat matplotlib histogram barlarına özel işlem yapılabilir:
for rect in ax.patches:
    height = rect.get_height()
    if height > 0:
        ax.annotate(f'{int(height)}',
                    (rect.get_x() + rect.get_width() / 2., height),
                    ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")

# Ortalama kredi tutarını kredi notlarına göre gösteren çubuk grafik
ax = sns.barplot(data=df, x='loan_grade', y='loan_amnt', estimator='mean', palette='coolwarm')

# Başlık ve eksen etiketleri
plt.title("Kredi Notuna Göre Ortalama Kredi Tutarı", fontsize=14)
plt.xlabel("Kredi Notu", fontsize=12)
plt.ylabel("Ortalama Tutar ($)", fontsize=12)

# Her bar üzerine değer yazma
for p in ax.patches:
    height = p.get_height()
    if not np.isnan(height):
        ax.annotate(f'{height:.0f}',
                    (p.get_x() + p.get_width() / 2., height),
                    ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
categories = ['person_age', 'person_income', 'loan_amnt', 'loan_int_rate', 'person_emp_length']

group_means = df.groupby('loan_grade')[categories].mean()

scaler = MinMaxScaler()
normalized_values = scaler.fit_transform(group_means)
normalized_df = pd.DataFrame(normalized_values, columns=categories, index=group_means.index)

labels = categories
num_vars = len(labels)

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(5, 5), subplot_kw=dict(polar=True))

for grade in normalized_df.index:
    values = normalized_df.loc[grade].tolist()
    values += values[:1]

    ax.plot(angles, values, label=f'Not: {grade}', linewidth=2)
    ax.fill(angles, values, alpha=0.25)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=12)

ax.set_yticklabels([])

plt.title("Kredi Notlarına Göre Ortalama Değerler (Normalize Edilmiş Radar Grafiği)", fontsize=14, fontweight='bold', pad=20)
plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1.1))
plt.tight_layout()
plt.show()
```

```python
df['age_group'] = pd.cut(df['person_age'], bins=[18,25,35,45,55,65,100],
                         labels=["18-25", "26-35", "36-45", "46-55", "56-65", "65+"] )

age_approval = df[df['loan_status'] == 1]['age_group'].value_counts()
age_total = df['age_group'].value_counts()

approval_rate = (age_approval / age_total * 100).sort_index()
approval_rate = approval_rate.dropna()

age_labels = ["18-25", "26-35", "36-45", "46-55", "56-65", "65+"]
approval_rate = approval_rate.reindex(age_labels, fill_value=0)

plt.figure(figsize=(10,6))
ax = sns.barplot(x=approval_rate.index, y=approval_rate.values, color="steelblue")

plt.title("Yaş Aralığına Göre Kredi Onay Oranı (%)")
plt.ylabel("Onay Oranı (%)")
plt.xlabel("Yaş Grubu")

for i, v in enumerate(approval_rate.values):
    ax.text(i, v + 1, f"{v:.1f}%", ha='center', fontsize=11, color='black')

ymax = max(approval_rate.values)
plt.ylim(0, ymax * 1.15 if ymax > 0 else 1)

plt.show()
```
